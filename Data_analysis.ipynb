```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris_data = load_iris()

# Create a pandas DataFrame from the dataset
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the 'species' column from the target data
df['species'] = iris_data.target_names[iris_data.target]

print("Dataset loaded successfully!")
```

```python
print("First 5 rows of the dataset:")
df.head()
```

```python
print("Dataset information:")
df.info()
```

```python
print("Database is clean and ready for analysis.")
```

```python
# Basic data analysis
print("Statistical summary of the dataset:")
print(df.describe())
```

```python
#  Statistical analysis by species
mean = df.groupby("species").mean()
max = df.groupby("species").max()
min = df.groupby("species").min()

# Display the results
print("Mean values od species:")
print(mean)
print("Max values of species:")
print(max)
print("Min values of species:")
print(min)
```

```python
# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns   

sns.set_style("whitegrid")

# Line Graph
plt.figure(figsize=(10, 6))
plt.plot(df['sepal length (cm)'])
plt.title("Sepal Length line Graph")
plt.xlabel("index")
plt.ylabel("sepal length (cm)")
plt.show()
```

```python
# bar plot
plt.figure(figsize=(8, 6))
mean["petal length (cm)"].plot(kind ="bar", color = ['blue', 'orange', 'green'])
plt.title('Average petal length by species')
plt.xlabel('Species')
plt.ylabel('Average petal length (cm)')
plt.xticks(rotation=0) # Keep x-axis labels horizontal
plt.show()
```

```python
# Histogram with KDE using seaborn
plt.figure(figsize=(8, 6))
sns.histplot(df['petal width (cm)'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# scatterplot of sepal length vs sepal width
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='bright', s=100)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```
